```python
# LEGB → local, Enclosing, Global, Built-in
```

```python
isim = "Atıl"

def ornek_fonksiyon():
    #enclosing
    isim = "Atıl 2"
    print(isim)

    def ic_fonksiyon():
        #local
        isim = "Atıl 3"
        print(isim)

    ic_fonksiyon()    
```

```python
ornek_fonksiyon()
```
