```python
class Kullanici:

    meslek = ""

    def __init__(self, isim, yas, cinsiyet):
        self.isim = isim
        self.yas = yas
        self.cinsiyet = cinsiyet

    def ismi_yazdir(self):
        print(self.isim)
```

```python
atıl = Kullanici("atıl", 50, "erkek")
```

```python
atıl.ismi_yazdir()
```

```python
atıl.isim
```

```python
atıl.yas
```

```python
atıl.meslek = "yazılımcı"
```

```python
atıl.meslek
```

```python
class Kopek():
    
    tuy_rengi = "sarı"
    
    def __init__(self, isim, yas=0):
        self.isim = isim
        self.yas = yas
        self.insan_yasi = self.yas * 7

    def tuy_rengini_yazdir(self):
        print(self.tuy_rengi)   # → sel.tuy_rengi yerine → Kopek.tuy_rengi de yazılabilir
```

```python
barley = Kopek("Barley",7)
```

```python
barley.yas
```

```python
barley.insan_yasi
```

```python
barley.tuy_rengini_yazdir()
```
