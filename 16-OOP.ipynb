```python
class Muzisyen():

    def __init__(self,isim,yas):
        self.isim = isim
        self.yas = yas
        print("muzisyen sınıfından bir obje oluşturuldu")

    def ornek1(self):
        print("örnek 1")

    def ornek2(self):
        print("örnek 2")
```

```python
atil = Muzisyen("Atıl", 38)
```

```python
atil.ornek1()
```

```python
class OzelMuzisyen(Muzisyen):
    
    def __init__(self, isim, yas):
        Muzisyen.__init__(self, isim, yas)
        print("Özel Muzisyen sınıfından bir obje oluşturuldu")

    def ornek3(self):
        print("örnek 3")

    #override
    def ornek1(self):
        print("örnek 1 çalıştırıldı ama özel müzisyen")
```

```python
atlas = OzelMuzisyen("Atlas", 5)
```

```python
atlas.ornek1()
```

```python
atlas.ornek2()
```

```python
atlas.ornek3()
```

```python
# polymorphism
```

```python
class Muz:

    def __init__(self, isim):
        self.isim = isim

    def bilgi_ver(self):
        return f"{self.isim} 100 kaloridir"

class Elma:

    def __init__(self, isim):
        self.isim = isim

    def bilgi_ver(self):
        return f"{self.isim} 150 kaloridir"
```

```python
muz = Muz("muz")
```

```python
muz.bilgi_ver()
```

```python
elma = Elma("elma")
```

```python
elma.bilgi_ver()
```

```python
meyveler = [muz, elma]
```

```python
for meyve in meyveler:
    print(meyve.bilgi_ver())
```

```python
#encapsulation
```

```python
class Telefon:

    def __init__(self, isim, fiyat):
        self.__isim = isim
        self.fiyat = fiyat

    def bilgi_ver(self):
        print(f"{self.__isim} in fiyatı: {self.fiyat}")

    def isim_degistir(self, isim):
        self.__isim = isim

    def __gizli_method(self):
        print("gizli method çalıştırıldı")

    def test_methodu(self):
        self.__gizli_method()
```

```python
pixel = Telefon("Pixel XL", 500)
```

```python
pixel.bilgi_ver()
```

```python
pixel.isim = "Pixel XXL"
```

```python
pixel.bilgi_ver()
```

```python
pixel.isim_degistir("Pixel XXL")
```

```python
pixel.bilgi_ver()
```

```python
pixel.test_methodu()
```

```python
from abc import ABC, abstractmethod
```

```python
class Araba (ABC):

    @abstractmethod
    def max_hiz(self):
        pass

    @abstractmethod
    def beygir_gucu(self):
        pass
```

```python
class Ornek():

    def ornek(self):
        pass
```

```python
ornek = Ornek()
```

```python
class Mercedes(Araba):

    def __init__(self,isim):
        self.isim = isim

    def beygir_gucu(self):
        return 500

    def max_hiz(self):
        return 100
```

```python
merc = Mercedes("mercedes c 200")
```

```python
merc.beygir_gucu()
```

```python
merc.max_hiz()
```

```python
class Meyve:

    def __init__(self, isim, kalori):
        self.isim = isim
        self.kalori = kalori

    def __str__(self):
        return f"{self.isim} 'ın kalorisi şudur: {self.kalori}"

    def __len__(self):
        return self.kalori
```

```python
portakal = Meyve("portakal", 200)
```

```python
print(portakal)
```

```python
len(portakal)
```

```python
uzum = Meyve("üzüm", 250)
```

```python
len(uzum)
```
