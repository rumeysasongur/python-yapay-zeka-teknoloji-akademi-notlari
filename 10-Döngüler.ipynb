```python
# continue, break, pass
```

```python
numara_listesi = [10,20,30,40,50,60,70,80]
```

```python
print("döngü başladı")
for num in numara_listesi:
    print(num)
print("döngü bitti")    
```

```python
for num in numara_listesi:
    print(num)
    if num == 40:
        break
```

```python
for num in numara_listesi:
    if num ==40:
        continue
    print(num)    
```

```python
if 4 > 3:
    pass
```

```python
for num in numara_listesi:
    pass
```

```python
x = 0
print("döngü başladı")
while x < 10:
    print(x)
    x = x + 1
print("döngü bitiyor")    
```
